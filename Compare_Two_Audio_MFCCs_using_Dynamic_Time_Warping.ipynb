```python
import librosa
#import librosa
import IPython.display as ipd
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import IPython
from scipy.io import wavfile
import scipy.signal
```

```python
Mi_off,misr = librosa.load('./output/redmi/cal_off/1.wav')
```

```python
Nova_off,novasr = librosa.load('./output/nova3i/cal_off/1.wav')
```

```python
samasung,sasr = librosa.load('./output/samsung/cal_off/1.wav')
```

```python
%pylab inline

subplot(1, 2, 1)
mfcc1 = librosa.feature.mfcc(Mi_off, misr)
librosa.display.specshow(mfcc1)

subplot(1, 2, 2)
mfcc2 = librosa.feature.mfcc(Nova_off, novasr)
librosa.display.specshow(mfcc2)
```

```python

from dtw import dtw
```

```python
from numpy.linalg import norm
dist, cost, acc_cost, path = dtw(mfcc1.T, mfcc2.T, dist=lambda x, y: norm(x - y, ord=1))
print('Normalized distance between the two sounds:', dist)
```
