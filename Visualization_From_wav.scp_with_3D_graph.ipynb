# Create Male and Female Information From wav.scp

```python
Male_wav = []
Female_wav =[]
below_20_l = []
below_30_l =[]
below_40_l =[]
above_40_l =[]
M_info = []
F_info = []
for i in data:
    Gender = i.split('/')[3].split('-')[2]
    Age = i.split('/')[3].split('-')[3]
    int_age = int(Age)
    Speaker = i.split('/')[2]
    Native_language = i.split('/')[3].split('-')[4]
    if Gender == 'M':
        Male = i + ': ' + i.split('/')[2] + ': ' + i.split('/')[3].split('-')[1]
        Male_wav.append(Male)
        M_info.append(i.split('/')[2])
        
        #set(Male_l)
    #print("Number of male:",len(Male_l))
    elif Gender == 'F':
        Female = i + ': ' + i.split('/')[2] + ': ' + i.split('/')[3].split('-')[1]
        Female_wav.append(Female)
        F_info.append(i.split('/')[2])
        #set(Female_l)
    #print("Number of female:",len(Female_l))
```

# Create each different Ages group from wav.scp

```python
below_20_l = []
below_30_l =[]
below_40_l =[]
above_40_l =[]
below_20_info = []
below_30_info = []
below_40_info = []
above_40_info = []
for i in data:
    Gender = i.split('/')[3].split('-')[2]
    Age = i.split('/')[3].split('-')[3]
    int_age = int(Age)
    Speaker = i.split('/')[2]
    Native_language = i.split('/')[3].split('-')[4]
    if int_age < 20:
        #print("I'm here")
        below_20 = i + ':' + i.split('/')[2] + '|' + i.split('/')[3].split('-')[1] + '|' + Age
        below_20_l.append(below_20)
        below_20_info.append(i.split('/')[2]+'|'+Age+'|'+Gender)
    elif 20 < int_age < 30:
        below_30 = i + ':' + i.split('/')[2] + '|' + i.split('/')[3].split('-')[1] + '|' + Age
        below_30_l.append(below_30)
        below_30_info.append(i.split('/')[2]+'|'+Age+'|'+Gender)
    elif 30 < int_age < 40:
        below_40 = i + ':' + i.split('/')[2] + '|' + i.split('/')[3].split('-')[1] + '|' + Age
        below_40_l.append(below_40) 
        below_40_info.append(i.split('/')[2]+'|'+Age+'|'+Gender)
    elif 40 < int_age:
        above_40 = i + ':' + i.split('/')[2] + '|' + i.split('/')[3].split('-')[1] + '|' + Age
        above_40_l.append(above_40)   
        above_40_info.append(i.split('/')[2]+'|'+Age+'|'+Gender)
```
